```python
All_Data = pd.read_csv("ALL_DATA.csv", skiprows=2, names=['Order ID','Product','Quantity','Price Each','Order Date','Purchase Adress'])
All_Data.head()
```

### I concatinate the data by typing them one by one so i can easily customize the concatinated data quarterly, yearly, or even some specific months, tho there is other way to concat them from one folder

# Cleanup The Data!

### Get rid of NaN

```python
nan_df = All_Data[All_Data.isna().any(axis=1)]
nan_df.head()

All_Data = All_Data.dropna(how='all')
All_Data.head()
```

### Get rid of 'Or' in two first string in order date column

```python
All_Data = All_Data[All_Data['Order Date'].str[0:2] != 'Or']
All_Data.head()
```

### Convert column to a correct type

```python
All_Data['Quantity'] = pd.to_numeric(All_Data['Quantity'])
All_Data['Price Each'] = pd.to_numeric(All_Data['Price Each'])

All_Data.head()
```

# Add a Month Column

```python
All_Data['Month'] = All_Data['Order Date'].str[0:2]
All_Data['Month'] = All_Data['Month'].astype('int32')
All_Data.head()
```

# Add a Sales Column

```python
All_Data['Sales'] = All_Data['Quantity'] * All_Data['Price Each']
All_Data.head()
```

# Add City Column

```python
All_Data['City'] = All_Data['Purchase Adress'].apply(lambda x: x.split(',')[1]) + (" ") + All_Data['Purchase Adress'].apply(lambda x: x.split(',')[2].split(' ')[1])
All_Data.head()
All_Data.head()
```

# Add State Column

```python
All_Data['State'] = All_Data['Purchase Adress'].apply(lambda x: x.split(',')[2].split(' ')[1])
All_Data.head()
```
